```python
import pandas as pd
#df = pd.read_csv("E:\DAIICT\sem 2\IT504(Distributed Database)\Project\weatherHistory.csv",index_col=False)
import psycopg2

conn = psycopg2.connect(database="202011060_db", user = "postgres", password = "admin", host = "localhost", port = "5432",options="-c search_path=project")
cur = conn.cursor()
```

```python
# one attr in group by, one aggregate function (constraints)
#query = 'SELECT Summary, avg(Temperature) from "Weather" where Precip_Type='+"'rain'"+' group by Summary'
query = 'select BrewMethod,avg(BoilGravity) from "Recipe" group by BrewMethod'
query
cur.execute(query)
ans=cur.fetchall()

import re
query_words = query.split()

index= query_words.index('group')
group_by_keyword=query_words[index+2]
ind = query_words.index('from')
table_name=query_words[ind +1]
if 'where' in query_words:
    i_where = query_words.index('where')
    list1 = query_words[i_where:index]
    query_part = " ".join(list1)
    query1 = 'select * from '+ table_name + query_part
else:
    query1 = 'select * from ' + table_name
    
cur.execute(query1)
tupples = cur.fetchall()
df = pd.DataFrame(tupples, columns=['BeerID','Name','URL','Style','StyleID','Size','OG','FG','ABV','IBU','Color','BoilSize','BoilTime','BoilGravity','Efficiency','MashThickness','SugarScale','BrewMethod','PitchRate','PrimaryTemp','PrimingMethod','PrimingAmount','UserId'
])
cur.close()
df
```

```python
dict_data = df.groupby(group_by_keyword).groups
#create dictionary for labels  
result = re.search(r"\(([A-Za-z0-9_]+)\)", query)
val = result.group(1)
val
```

```python
label_dict={}
for i in range(len(ans)):
    if ans[i][1]==None:
        label_dict[ans[i][0]] = "missing"
incorrect_missing = {}
for i in dict_data:
    if i not in label_dict:
        label_dict[i] = "correct"
        for j in range(len(dict_data[i])):   
            if df.loc[[dict_data[i][j]]][val].isna().values[0]:
                label_dict[i] = "incorrect"
                incorrect_missing[i] = "incorrect"
                break            
```
